###a. Defina en una variable el diccionario que creó en el Ejercicio 3 de la Tarea 0 - Reconocimiento de las herramientas y programación básica con Python.

**Ejercicio 3 de la Tarea 0**

 Realice un código para implementar la siguiente tarea: Construya un diccionario en el cual las llaves (keys) sean los nombres (strings) de ocho (8) personas (inventadas) de su entorno familiar o amigos y los valores (values) sean tuplas de cuatro (4) elementos que contengan la siguiente información de cada persona en orden: (Edad, Peso, Estatura, Estudia). Estos datos deben cumplir:

 • La edad debe ser en años.

 • El peso debe ser en kilogramos.

 • La estatura debe ser en metros.

 • El elemento “estudia” debe ser un Booleano, en el cual True significa que si estudia y False el caso contrario.

**Conceptos:**
Los diccionarios son estructuras de datos fundamentales en Python que permiten almacenar pares clave-valor. En este ejercicio, utilizaremos un diccionario para representar información de personas, donde los nombres serán las claves y los datos personales se almacenarán como tuplas en los valores. Las tuplas son secuencias inmutables que pueden contener elementos de diferentes tipos, lo que las hace ideales para agrupar datos relacionados.

**Explicación paso a paso:**

1.   Creamos un diccionario vacío llamado personas para almacenar la información.
2.   Definimos funciones de validación para edad, peso y estatura para asegurar que los datos ingresados sean válidos.
3.  Creamos una función agregar_persona que valida los datos y los agrega al diccionario si son correctos.
4.  Utilizamos la función agregar_persona para añadir 8 personas al diccionario con datos inventados.
5.  Finalmente, iteramos sobre el diccionario para imprimir la información de cada persona de manera formateada.


```python
# Creamos un diccionario vacío para almacenar la información de las personas
personas = {}

# Función para validar la edad (debe ser un número entero positivo)
def validar_edad(edad):
    return isinstance(edad, int) and edad > 0

# Función para validar el peso (debe ser un número positivo)
def validar_peso(peso):
    return isinstance(peso, (int, float)) and peso > 0

# Función para validar la estatura (debe ser un número positivo entre 0.5 y 2.5 metros)
def validar_estatura(estatura):
    return isinstance(estatura, float) and 0.5 <= estatura <= 2.5

# Función para agregar una persona al diccionario
def agregar_persona(nombre, edad, peso, estatura, estudia):
    if validar_edad(edad) and validar_peso(peso) and validar_estatura(estatura):
        personas[nombre] = (edad, peso, estatura, estudia)
    else:
        print(f"Datos inválidos para {nombre}. No se ha agregado.")

# Agregamos 8 personas al diccionario
agregar_persona("Ana", 25, 60.5, 1.65, True)
agregar_persona("Carlos", 30, 75.0, 1.80, False)
agregar_persona("Elena", 22, 55.5, 1.70, True)
agregar_persona("David", 35, 80.2, 1.75, False)
agregar_persona("Sofia", 28, 58.7, 1.68, True)
agregar_persona("Miguel", 40, 70.8, 1.78, False)
agregar_persona("Laura", 33, 62.3, 1.72, True)
agregar_persona("Pedro", 27, 68.9, 1.82, False)

# Imprimimos el diccionario resultante
for nombre, datos in personas.items():
    edad, peso, estatura, estudia = datos
    print(f"{nombre}: Edad: {edad} años, Peso: {peso:.1f} kg, Estatura: {estatura:.2f} m, Estudia: {estudia}")
```

### c. Defina una función que reciba como parámetros una lista y regrese (return) el promedio de dicha lista. Utilice dicha función para imprimir (print) el promedio de las listas del inciso anterior.

 **Explicación paso a paso:**
1.  Definimos el diccionario personas con la información de 8 personas, tal como se creó en el Ejercicio 3 de la Tarea 0.
2.  Creamos cuatro listas vacías: edades, pesos, estaturas y estudia.
3.  Utilizamos un ciclo for para iterar sobre los valores del diccionario personas. En cada iteración:
* Extraemos los datos de cada persona.
* Agregamos cada dato a la lista correspondiente usando el método append().
4.  Imprimimos las listas generadas para verificar que la extracción se realizó correctamente.
5.  Definimos la función calcular_promedio() que toma una lista como parámetro:
* Verificamos si la lista está vacía para evitar divisiones por cero.
* Utilizamos las funciones sum() y len() para calcular el promedio.
* Retornamos el resultado.
6.  Utilizamos la función calcular_promedio() para calcular e imprimir el promedio de cada lista.


```python
# a. Definimos el diccionario creado en el Ejercicio 3 de la Tarea 0
personas = {
    "Ana": (25, 60.5, 1.65, True),
    "Carlos": (30, 75.0, 1.80, False),
    "Elena": (22, 55.5, 1.70, True),
    "David": (35, 80.2, 1.75, False),
    "Sofia": (28, 58.7, 1.68, True),
    "Miguel": (40, 70.8, 1.78, False),
    "Laura": (33, 62.3, 1.72, True),
    "Pedro": (27, 68.9, 1.82, False)
}

# b. Implementamos ciclos para extraer la información y generar las listas
edades = []
pesos = []
estaturas = []
estudia = []

for datos in personas.values():
    edades.append(datos[0])
    pesos.append(datos[1])
    estaturas.append(datos[2])
    estudia.append(datos[3])

# Imprimimos las listas para verificar
print("Edades:", edades)
print("Pesos:", pesos)
print("Estaturas:", estaturas)
print("Estudia:", estudia)

# c. Definimos una función para calcular el promedio de una lista
def calcular_promedio(lista):
    """
    Calcula el promedio de los elementos en una lista.

    Parámetros:
    lista (list): Una lista de números.

    Retorna:
    float: El promedio de los números en la lista.
    """
    if not lista:  # Verificamos si la lista está vacía
        return 0
    return sum(lista) / len(lista)

# Calculamos e imprimimos los promedios
print("\nPromedios:")
print("Edad promedio:", calcular_promedio(edades))
print("Peso promedio:", calcular_promedio(pesos))
print("Estatura promedio:", calcular_promedio(estaturas))
print("Proporción de estudiantes:", calcular_promedio(estudia))
```

Esto nos permitirá organizar nuestros DataFrames de manera eficiente y acceder a ellos fácilmente.

**Explicación paso a paso:**

* Programe una función que extraiga todas las listas de los encabezados de columnas que se encuentran en un Panel y úsela para extraer las listas de columnas del Panel del punto anterior. Imprima los resultados de forma que se diferencien los encabezados entre los dataframes en el Panel.

**Explicación paso a paso:**
1.  Definimos la función extraer_encabezados_panel(panel) que toma como argumento nuestro Panel (diccionario de DataFrames) y devuelve un diccionario donde las claves son los nombres de los DataFrames y los valores son listas de los encabezados de columnas.
2.  Utilizamos esta función para extraer los encabezados de nuestro Panel y almacenarlos en encabezados_panel.
3.  Imprimimos los resultados de forma estructurada:
  * Mostramos los encabezados de cada DataFrame por separado, numerándolos para mayor claridad.
  * Utilizamos separadores para diferenciar visualmente los encabezados de cada DataFrame.
4.  Proporcionamos información adicional:
  * Mostramos un resumen del número de columnas en cada DataFrame.
  * Buscamos e imprimimos las columnas que son comunes a todos los DataFrames (si las hay).

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Asumimos que ya tenemos cargados df1, df2 y df3 del ejercicio anterior

# Crear el Panel (diccionario de DataFrames)
panel = {
    "Excel_Data": df1,
    "CSV_Data": df2,
    "Breast_Cancer_Data": df3
}

def extraer_encabezados_panel(panel):
    """
    Extrae las listas de encabezados de columnas de un Panel.

    Args:
    panel (dict): Un diccionario donde las claves son nombres de DataFrames
                  y los valores son los DataFrames correspondientes.

    Returns:
    dict: Un diccionario donde las claves son los nombres de los DataFrames
          y los valores son listas de los encabezados de columnas.
    """
    encabezados = {}
    for nombre, df in panel.items():
        encabezados[nombre] = df.columns.tolist()
    return encabezados

# Usar la función para extraer los encabezados
encabezados_panel = extraer_encabezados_panel(panel)

# Imprimir los resultados
print("Encabezados de columnas en cada DataFrame del Panel:")
for nombre, columnas in encabezados_panel.items():
    print(f"\n{nombre}:")
    for i, columna in enumerate(columnas, 1):
        print(f"  {i}. {columna}")
    print("-" * 50)

# Información adicional sobre los encabezados
print("\nResumen de encabezados:")
for nombre, columnas in encabezados_panel.items():
    print(f"{nombre}: {len(columnas)} columnas")

# Buscar columnas comunes (si las hay)
todos_encabezados = set.intersection(*[set(cols) for cols in encabezados_panel.values()])
if todos_encabezados:
    print("\nColumnas comunes en todos los DataFrames:")
    for columna in todos_encabezados:
        print(f"  - {columna}")
else:
    print("\nNo hay columnas comunes en todos los DataFrames.")
```

* Programe una función que imprima la cantidad de elementos que tienen las bases de datos en un Panel y úsela con el Panel del punto anterior. Imprima los resultados de forma que se diferencien los tamaños de los diferentes dataframes en el Panel.

**Explicación paso a paso:**
1.  Definimos la función imprimir_tamanios_panel(panel) que toma como argumento nuestro Panel (diccionario de DataFrames) e imprime la cantidad de elementos en cada DataFrame.
2.  En la función:
  * Calculamos el número de filas, columnas y elementos totales para cada DataFrame.
  * Formateamos la salida para que esté alineada y sea fácil de leer, usando justificación a la izquierda para los nombres y separadores visuales.
  * Calculamos y mostramos el total de elementos en todo el Panel.
3.  Utilizamos esta función para imprimir los tamaños de los DataFrames en nuestro Panel.
4.  Proporcionamos información adicional útil:
  * Mostramos los tipos de datos en cada DataFrame.
  * Calculamos y mostramos la memoria utilizada por cada DataFrame.
5.  Identificamos y mostramos cuál es el DataFrame más grande y cuál el más pequeño en términos de número de elementos.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Asumimos que ya tenemos cargados df1, df2 y df3 como en ejercicios anteriores

# Crear el Panel (diccionario de DataFrames)
panel = {
    "Excel_Data": df1,
    "CSV_Data": df2,
    "Breast_Cancer_Data": df3
}

def imprimir_tamanios_panel(panel):
    """
    Imprime la cantidad de elementos en cada DataFrame de un Panel.

    Args:
    panel (dict): Un diccionario donde las claves son nombres de DataFrames
                  y los valores son los DataFrames correspondientes.
    """
    print("Tamaños de los DataFrames en el Panel:")
    print("=" * 50)

    total_elementos = 0
    max_nombre_len = max(len(nombre) for nombre in panel.keys())

    for nombre, df in panel.items():
        filas, columnas = df.shape
        elementos = filas * columnas
        total_elementos += elementos

        print(f"{nombre.ljust(max_nombre_len)} : {filas:,} filas x {columnas:,} columnas = {elementos:,} elementos")
        print("-" * 50)

    print(f"{'Total'.ljust(max_nombre_len)} : {total_elementos:,} elementos en todo el Panel")
    print("=" * 50)

# Usar la función para imprimir los tamaños
imprimir_tamanios_panel(panel)

# Información adicional sobre los DataFrames
print("\nInformación adicional sobre los DataFrames:")
for nombre, df in panel.items():
    print(f"\n{nombre}:")
    print(f"  - Tipos de datos:\n{df.dtypes}")
    print(f"  - Memoria utilizada: {df.memory_usage().sum() / 1024:.2f} KB")
    print("-" * 50)

# Identificar el DataFrame más grande y más pequeño
df_sizes = {nombre: df.size for nombre, df in panel.items()}
largest_df = max(df_sizes, key=df_sizes.get)
smallest_df = min(df_sizes, key=df_sizes.get)

print(f"\nDataFrame más grande: {largest_df} con {df_sizes[largest_df]:,} elementos")
print(f"DataFrame más pequeño: {smallest_df} con {df_sizes[smallest_df]:,} elementos")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asumimos que df2 ya está cargado de ejercicios anteriores
# Si no, lo cargaríamos así:
# df2 = pd.read_csv('Anexo 2 - Base de Datos CSV.csv')

# Paso 1: Mostrar información general del DataFrame
print("Información general del DataFrame df2:")
print(df2.info())
print("\nPrimeras filas del DataFrame df2:")
print(df2.head())

# Paso 2: Contar los elementos nulos en cada columna
nulos_por_columna = df2.isnull().sum()

# Paso 3: Mostrar el conteo de elementos nulos
print("\nCantidad de elementos nulos por columna:")
print(nulos_por_columna)

# Paso 4: Calcular el porcentaje de nulos
porcentaje_nulos = (nulos_por_columna / len(df2)) * 100

# Paso 5: Mostrar el porcentaje de elementos nulos
print("\nPorcentaje de elementos nulos por columna:")
print(porcentaje_nulos)

# Paso 6: Crear un DataFrame con el conteo y porcentaje de nulos
resumen_nulos = pd.concat([nulos_por_columna, porcentaje_nulos], axis=1,
                          keys=['Conteo de Nulos', 'Porcentaje de Nulos'])
resumen_nulos = resumen_nulos.sort_values('Conteo de Nulos', ascending=False)

print("\nResumen de elementos nulos:")
print(resumen_nulos)

# Paso 7: Visualización gráfica de los nulos
plt.figure(figsize=(12, 6))
resumen_nulos['Conteo de Nulos'].plot(kind='bar')
plt.title('Cantidad de Valores Nulos por Columna')
plt.xlabel('Columnas')
plt.ylabel('Número de Valores Nulos')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Paso 8: Información adicional sobre el DataFrame
print("\nEstadísticas descriptivas del DataFrame df2:")
print(df2.describe())

print("\nTipos de datos de las columnas:")
print(df2.dtypes)

# Paso 9: Sugerir acciones basadas en el análisis
print("\nSugerencias basadas en el análisis:")
for columna, nulos in nulos_por_columna.items():
    if nulos > 0:
        porcentaje = porcentaje_nulos[columna]
        if porcentaje > 50:
            print(f"- La columna '{columna}' tiene {nulos} valores nulos ({porcentaje:.2f}%). "
                  f"Considere eliminar esta columna o imputar los valores faltantes.")
        elif porcentaje > 10:
            print(f"- La columna '{columna}' tiene {nulos} valores nulos ({porcentaje:.2f}%). "
                  f"Se recomienda investigar la razón de estos valores faltantes y considerar imputarlos.")
        else:
            print(f"- La columna '{columna}' tiene {nulos} valores nulos ({porcentaje:.2f}%). "
                  f"Estos pueden ser manejados mediante imputación o eliminación de filas, dependiendo del contexto.")
```

```python
# Paso 1: Importación de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Paso 2: Cargar los datos del archivo Excel
df1 = pd.read_excel('https://github.com/aisergio/Python/blob/main/Anexo1-Base_de_Datos-EXCEL.xlsx?raw=true',
                    sheet_name=1,  # Segunda hoja (índice 1)
                    header=3,      # La fila 4 contiene los encabezados
                    index_col=0)   # Primera columna como índice

# Paso 3: Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Paso 4: Crear los histogramas
ax1.hist(df1['Tiempo motor ON (h)'], bins=30, alpha=0.5, label='Tiempo motor ON', color='blue')
ax1.hist(df1['Tiempo en Movimiento (h)'], bins=30, alpha=0.5, label='Tiempo en Movimiento', color='red')

# Configurar el primer subplot (histogramas)
ax1.set_title('Distribución de Tiempos de Motor y Movimiento', fontsize=16)
ax1.set_xlabel('Tiempo (horas)', fontsize=12)
ax1.set_ylabel('Frecuencia', fontsize=12)
ax1.legend(fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.7)

# Paso 5: Crear el gráfico de dispersión
sns.scatterplot(x='Tiempo motor ON (h)', y='Tiempo en Movimiento (h)', data=df1, ax=ax2)

# Configurar el segundo subplot (gráfico de dispersión)
ax2.set_title('Relación entre Tiempo motor ON y Tiempo en Movimiento', fontsize=16)
ax2.set_xlabel('Tiempo motor ON (horas)', fontsize=12)
ax2.set_ylabel('Tiempo en Movimiento (horas)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)

# Paso 6: Añadir línea de tendencia usando regresión lineal
x = df1['Tiempo motor ON (h)']
y = df1['Tiempo en Movimiento (h)']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept
ax2.plot(x, line, color='red', linestyle='--', label=f'y = {slope:.4f}x + {intercept:.4f}')
ax2.legend(fontsize=10)

# Paso 7: Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()

# Paso 8: Análisis estadístico
print("Estadísticas descriptivas:")
print(df1[['Tiempo motor ON (h)', 'Tiempo en Movimiento (h)']].describe())

print(f"\nCorrelación entre Tiempo motor ON y Tiempo en Movimiento: {r_value:.4f}")
print(f"Ecuación de la línea de tendencia: y = {slope:.4f}x + {intercept:.4f}")
print(f"R-cuadrado: {r_value**2:.4f}")
print(f"Valor p: {p_value:.4f}")
```
